## lan variable

```r
library("randomForest")
cellNumber <- 257
load(paste("../2_variables/frames",as.character(cellNumber), "cells_frame.RData", sep="/"))
set.seed(1)
```

```r
drosophila.df <- cells_frame
cond <- drosophila.df$type == 'dWL' | drosophila.df$type == 'dWP'
data.class <- droplevels(drosophila.df[cond,])
drosophila.df <- drosophila.df[cond,]
```

```r
training.dWL <- c()
training.dWP <- c()
training.all <- c()

validation.dWL <- c()
validation.dWP <- c()
validation.all <- c()

for (i in seq(10000)){
    data_set_size <- floor(nrow(data.class)/4*3)

    indexes <- sample(1:nrow(data.class), size = data_set_size)

    training <- data.class[indexes,]
    validation1 <- data.class[-indexes,]

    rf <- randomForest(formula = type ~ lan.1.02.sup, data = training, ntree = 200)
    training.dWL <- c(training.dWL, rf$confusion['dWL',3])
    training.dWP <- c(training.dWP, rf$confusion['dWP',3])
    pm <- rf$confusion
    training.all <- c(training.all, 1 - (pm[1,1] + pm [2,2])/sum(pm))
    
    prediction_for_table <- predict(rf,validation1)
    pm <- table(observed=validation1[,'type'],predicted=prediction_for_table)
    validation.dWL <- c(validation.dWL, 1 - pm['dWL', 'dWL']/sum(pm['dWL',]))
    validation.dWP <- c(validation.dWP, 1 - pm['dWP', 'dWP']/sum(pm['dWP',]))
    validation.all <- c(validation.all, 1 - (pm[1,1] + pm [2,2])/sum(pm))
        
    }
```

```r
print('dWL acc')
1-mean(training.dWL)
print('dWP acc')
1-mean(training.dWP)
print('global acc')
1-mean(training.all)
```

```r
validation.dWL[is.nan(validation.dWL)] = 1
validation.dWP[is.nan(validation.dWP)] = 1
validation.all[is.nan(validation.all)] = 1

print('dWL acc')
1-mean(validation.dWL)
print('dWP error')
1-mean(validation.dWP)
print('global error')
1-mean(validation.all)
```
